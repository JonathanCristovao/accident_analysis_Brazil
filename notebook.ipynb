```python
import pandas as pd
import glob
import warnings
warnings.filterwarnings('ignore')
```

```python
arquivos_csv = glob.glob("data/*.csv")

dfs = []

for arquivo_csv in arquivos_csv:
    df = pd.read_csv(arquivo_csv, sep=';', encoding='latin-1', parse_dates=['data_inversa'])
    dfs.append(df)

resultado = pd.concat(dfs, ignore_index=True)

resultado.to_csv("concatenado.csv", index=False, sep=';') 
```
